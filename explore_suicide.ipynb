## Data Preparation and Cleaning

- Stage one **[Preprocessing the dataset]**

```python
import pandas as pd # importing the pandas library
import numpy as np # importing the numpy library
```

```python
suicide_df = pd.read_csv('master.csv') # loading the dataset usiung pandas
print('The number of rows and columns in the dataset -{}'.format(suicide_df.shape)) # Displays the shape/size of dataset
```

```python
suicide_df.head()
```

```python
suicide_df.head(100) # Displaying 0 - 100 of the dataset
```

##### dataset did not display complete so I used the **pd.set_option** to display 0 - 99 values in the dataset.

```python
suicide_df.head(100)
```

```python
suicide_df.head(10).isna()
```

```python
suicide_df.rename(columns={' gdp_for_year ($) ': 'gdp_yearly', 'gdp_per_capita ($)': 'gdp_per_capita', 'suicides/100k pop': 'suicides_100k_per_pop' }, inplace=True) # Renaming the columns makes it easier to use
```

```python
suicide_df.columns
```

```python
suicide_df['population'].head()
```

```python
suicide_df['gdp_yearly'].head()
```

```python
suicide_df['gdp_yearly'].sum()

# This line of code is not giving me my desired output because pandas 
# is reading 'gdp_yearly' as an object data type.
# The best way to get the desired output is by converting the object dtype to a floating/int type value
```

```python
suicide_df['gdp_per_population'] = suicide_df['gdp_yearly'] / suicide_df['population']

# This line of code is returning an error message because 'gdp_yearly' has not being converted to a numeric value
```

```python
suicide_df['gdp_yearly'] = suicide_df['gdp_yearly'].str.replace(',', '') # Removing comma from gdp_yearly column 
```

```python
suicide_df['gdp_yearly'].head().astype(float) # Now we convert the str type to float point value 
```

```python
suicide_df['gdp_per_population'] = suicide_df['gdp_yearly'].astype(float) / suicide_df['population'] # Now we can perform mathematical operations using the gdp_yearly column
# Creating a new column called gdp_per_population
```

## Scatter plot

```python
plt.figure(figsize=(12,6))
plt.title('Sexual dominance of people who commit suicide')
sns.scatterplot(suicide_df.suicides_no, suicide_df.population, hue=suicide_df.sex, s=100);
```

```python
plt.figure(figsize=(12,6))
plt.title('Age of people who commit suicide')
sns.scatterplot('suicides_no', 
                'population', 
                hue = 'age', 
                s = 100,
               data = suicide_df);
```

#### Question 2: Merging datasets

```python
country_df = pd.read_csv('Country.csv', usecols=['Country.Code', 'Currency.Unit'])
world_bank_df = pd.read_csv('world_bank.csv', skiprows=4, usecols=['Country Name', 'Country Code'])

world_bank_df.columns
```

```python
country_df.columns
```

```python
country_df.rename(columns={'Currency.Unit':'currency', 'Country.Code': 'country_code'}, inplace=True)
world_bank_df.rename(columns={'Country Name': 'country', 'Country Code': 'country_code'}, inplace=True)
```

```python
country_df.columns
```

```python
world_bank_df.columns
```

#### Lets give our suicide dataset a country_code and currency

```python
world_X_country = world_bank_df.merge(country_df, on='country_code')
```

```python
world_X_country
```

```python
suicide_df.head()
```

#### Then we merge on the country column.

```python
suicide_df = suicide_df.merge(world_X_country, on='country')
```

```python
suicide_df.head(100)
```

```python
suicide_df.shape
```
